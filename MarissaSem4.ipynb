##PROBABILIDAD Y ESTADISTICA

**MEDIDAS DE LOCALIZACION**

 Son conceptos estadísticos utilizados para describir la posición central de un conjunto de datos. Estas indican un valor representativo que resume o caracteriza la distribución de los datos.

**MEDIDAS DE TENDECIA CENTRAL**

1. Media (Promedio) Es la suma de todos los valores del conjunto de datos dividida entre el número total de elementos.

2. Mediana: Es el valor que ocupa la posición central cuando los datos están ordenados de menor a mayor. Si el número de datos es impar, es el valor central; si es par, es el promedio de los dos valores centrales.

3. Moda: Es el valor o los valores que más se repiten en un conjunto de datos. Puede haber más de una moda (bimodal, multimodal) o ninguna moda si todos los valores son diferentes.

**DATOS ORDENADOS**

Cuartiles: Dividen los datos en 4 partes (25% cada uno).
Q1: 25%, Q2 (mediana): 50%, Q3: 75%.

Deciles: Dividen los datos en 10 partes (10% cada uno).
D1: 10%, D2: 20%, ..., D9: 90%.

Percentiles: Dividen los datos en 100 partes (1% cada uno).
P1: 1%, P2: 2%, ..., P99: 99%.

**MEDIDAS DE DISPERSIÓN**

Rango: Diferencia entre el máximo y el mínimo.

Varianza: Promedio de las diferencias al cuadrado respecto a la media.

Desviación estándar: Raíz cuadrada de la varianza, en las mismas unidades que los datos.
